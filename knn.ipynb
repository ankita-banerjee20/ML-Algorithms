#### **Importing Necessary Libraries**

```python
from sklearn.datasets import make_classification
import numpy as np
from sklearn.model_selection import train_test_split
from statistics import mode
from sklearn.metrics import accuracy_score
```

#### **Dataset Generation**

```python
X, y = make_classification(n_samples=100, n_features=4, n_redundant=0, n_classes=2, n_clusters_per_class=1)
```

#### **Train-Test Split**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

#### **Model**

```python
class KNearestNeighbours():
    def __init__(self, k):
        self.k = k
    
    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def predict(self, X_test):
        self.X_test = X_test

        pred_label = []
        for i in self.X_test:
            distances = []
            for j in self.X_train:
                distances.append(np.linalg.norm(i - j))

            k_neighbours = sorted(list(enumerate(distances)), key=lambda x: x[1])[:self.k]

            votes = []
            for i in k_neighbours:
                votes.append(self.y_train[i[0]])
            pred_label.append(mode(votes))

        return pred_label
```

#### **Model Fit**

```python
model = KNearestNeighbours(5)
model.fit(X_train, y_train)
```

#### **Prediction**

```python
y_pred = model.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```
